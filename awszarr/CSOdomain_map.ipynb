```python
import numpy as np
import geopandas as gpd
from shapely import geometry as sgeom
import cartopy.feature as cfeature
import cartopy.crs as ccrs

import matplotlib.pyplot as plt
```

```python
#WY bounding box
Bbox = {
    'latmax' : 44.582480,
    'latmin' : 42.363116,
    'lonmax': -109.477849,
    'lonmin': -111.155208,}

# Convert the bounding box dictionary to a shapely Polygon geometry using sgeom.box
box_sgeom = sgeom.box(Bbox['lonmin'], Bbox['latmin'], Bbox['lonmax'], Bbox['latmax'])
box_gdf = gpd.GeoDataFrame(geometry=[box_sgeom])#, crs=states.crs)
box_gdf.plot()
```

```python
#states
states_provinces = cfeature.NaturalEarthFeature(
    category='cultural',
    name='admin_1_states_provinces_lines',
    scale='50m',
    facecolor='none')
```

```python
#states
countries = cfeature.NaturalEarthFeature(
    category='cultural',
    name='ne_10m_admin_0_countries',
    scale='50m',
    facecolor='none')
countries()
```

```python
# set projection
to_proj = ccrs.PlateCarree()

fig, ax = plt.subplots(figsize=(10,8))

# Using cartopy, create the map projection and plot the data
ax = plt.axes(projection=to_proj)
# ax.add_feature(states_provinces,edgecolor='grey',linestyle=':')
# ax.coastlines(color='dimgrey')
# ax.add_feature(cfeature.BORDERS,edgecolor='dimgrey')
#grey states
ax.add_feature(states_provinces,edgecolor='white',linewidth=.6)#,linestyle=':')
ax.add_feature(cfeature.BORDERS,edgecolor='white',linewidth=1.5)
ax.add_feature(cfeature.LAND, facecolor = 'silver')
#ax.add_feature(cfeature.OCEAN, alpha = .3)
box_gdf.plot(alpha = .8, facecolor='lime', edgecolor='lime',ax=ax)

plt.xlim(-127, -63)
plt.ylim(24, 50)
```
